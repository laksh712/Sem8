```python
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
```

```python
boston_dataset=load_boston()
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"  #error solution provided by jupyter itself
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
df = pd.DataFrame(boston_dataset.data,columns=boston_dataset.feature_names) #create dataframe
df['MEDV'] = boston_dataset.target  #set target variable i.e median value of owner occupied homes
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x=df.loc[:,df.columns!='MEDV']
y=df.loc[:,df.columns=='MEDV']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
x_train=mms.fit_transform(x_train)
x_test=mms.transform(x_test)
```

```python
#define our deep neural network
from keras.models import Sequential
from keras.layers import Dense

model=Sequential()

model.add(Dense(128,input_shape=(13,),activation='relu',name='dense_1')) #input layer with 128 neurons
model.add(Dense(64,activation='relu',name='dense_2')) #hidden layer with 64 neurons
model.add(Dense(1,activation='linear',name='deense_output')) #output layer

model.compile(optimizer='adam',loss='mse',metrics=['mae'])
model.summary()
```

```python
#training model with help of training data
history=model.fit(x_train,y_train,epochs=100,verbose=1)
```

```python
#evaluate model on testing data
mse,mae=model.evaluate(x_test,y_test)
print('MSE : ',mse)
print('MAE : ',mae)
```
